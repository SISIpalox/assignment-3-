```python
import random
import matplotlib.pyplot as plt

def play_game(numbers):
    while len(numbers) > 1:  # Continue until only one number remains
        # Randomly select two distinct indices
        i, j = random.sample(range(len(numbers)), 2)
        # Calculate the difference
        diff = abs(numbers[i] - numbers[j])
        # Remove the original two numbers
        numbers.pop(max(i, j))  # Remove higher index first to avoid index shifting
        numbers.pop(min(i, j))  # Then remove the lower index
        # Add the difference back to the list
        numbers.append(diff)
    return numbers[0]

# Number of simulations
num_simulations = 1000

# List to store the final numbers
final_numbers = []

# Simulate the game 1000 times
for _ in range(num_simulations):
    numbers = list(range(1, 2025))  # Initialize the list of numbers for each simulation
    final_number = play_game(numbers)
    final_numbers.append(final_number)

# Plotting the results
plt.hist(final_numbers, bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Final Numbers After 1000 Simulations')
plt.xlabel('Final Number')
plt.ylabel('Frequency')
plt.show()
```
