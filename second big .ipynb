```python
import random
import matplotlib.pyplot as plt

def play_game(numbers):
    while len(numbers) > 1:  # Continue until only one number remains
        # Randomly select two distinct indices
        i, j = random.sample(range(len(numbers)), 2)
        # Calculate the difference
        diff = abs(numbers[i] - numbers[j])
        # Remove the original two numbers
        numbers.pop(max(i, j))  # Remove higher index first to avoid index shifting
        numbers.pop(min(i, j))  # Then remove the lower index
        # Add the difference back to the list
        numbers.append(diff)
    return numbers[0]

# Adjusting the simulation to be less computationally intensive
num_simulations_adjusted = 1000  # Number of simulations
number_range_adjusted = 5000  # Number of papers

final_numbers_adjusted = []

# Simulate the game with adjusted parameters
for _ in range(num_simulations_adjusted):
    numbers = list(range(1, number_range_adjusted + 1))  # Initialize the list of numbers for each simulation
    final_number = play_game(numbers)
    final_numbers_adjusted.append(final_number)

# Plotting the results for the adjusted simulation parameters
plt.hist(final_numbers_adjusted, bins=50, color='purple', edgecolor='black')
plt.title('Distribution of Final Numbers After 1000 Simulations with 5000 Papers')
plt.xlabel('Final Number')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

def play_game(numbers):
   
    while len(numbers) > 1:
        
        idx = np.random.choice(len(numbers), 2, replace=False)
       
        diff = abs(numbers[idx[0]] - numbers[idx[1]])
  
        max_idx = max(idx)
        min_idx = min(idx)
        numbers.pop(max_idx)
        numbers.pop(min_idx)
        
        numbers.append(diff)
   
    return numbers[0] if numbers else 0


num_simulations_reduced = 5000
final_numbers_reduced = []


for _ in range(num_simulations_reduced):
    numbers = list(range(1, 2024 + 1))
    final_number = play_game(numbers)
    final_numbers_reduced.append(final_number)


plt.hist(final_numbers_reduced, bins=50, color='red', edgecolor='black', log=True)
plt.title('Distribution of Final Numbers After 5,000 Simulations with 2024 Papers')
plt.xlabel('Final Number')
plt.ylabel('Frequency')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
